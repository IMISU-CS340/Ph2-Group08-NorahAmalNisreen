# خوارزمية Alpha-Beta pruning : 

تتشابه خوارزمية Alpha-Beta pruning مع خوارزمية minimax في أن كلا الخوارزميتين تستطيعان إيجاد أفضل حركة للاعب عند نقطة معينة. تختلف خوارزمية Alpha-Beta pruning عن خوارزمية minimax  في أنها تستطيع إيجاد الحل بشكل أسرع، نظراً لأنها تستطيع حذف بعض التفرعات من شجرة البحث التي لن تؤثر على النتيجة النهائية.



تتمثل مساحة المشكلة للعبة ب شجرة البحث , حيث تمثل النقاط الحالة الحالية للعبة و تمثل المسارات الخيارات التي يمكن للاعب اختيارها . وأول نقطة فالشجرة هي الموضع الذي يجب اتخاذ قرار فيه (اختيار حركة) ومن ثم يتم حساب معادلة التقييم لكل النقاط في نهاية شجرة البحث او (عند عمق محدد اذا تم تحديده) وتمثل هذه النقاط كيف تنتهي اللعبة .
-  الأفضل ان تستمر شجرة البحث الى النهاية لتحديد الفوز , ولكن يتم وضع عمق محدد لتقليل الوقت المطلوب والمساحة المطلوبة من الذاكرة 
- معادلة التقييم : إ ذا كانت ل النقاط الاخيره للشجرة البحث فهي تحدد الفوز والخساره والتعادل , اما إذا كانت عند عمق محدد فأنة من خلال عدد يمكن التنبؤ  بنسبة  الفوز  للاعب في هذا الاختيار     

 ###  شرح آلية عمل خوارزمية (Minimax)

#### شرح مبسَط حتى تتمكن من فهم خوارزمية (Alpha-Beta pruning)

  الفكرة الأساسية : يختار اللاعب الاول افضل خيار متاح له باعتبار ان اللاعب الأخر يختار أفضل خيار متاح له ايضاً 
- يتم بناء شجرة اللعبة الى عمق محدد وهو H  
- يتم حساب معادلة التقييم عند كل عقدة في العمق H
- يتم ارجاع الناتج من معادلة التقييم المطبقة النقاط في العمق الى نقطة الجذر(أول نقطة في الشجرة)

  - تقوم عقدة Max بأختيار اعلى قيمة لمعادلة التقييم لأبنأها
  
  - تقوم عقدة ال Min باختيار اقل قيمة لمعادلة التقييم لأبنأها 
- الوقت المطلوب لخوارزمية minimax =  عدد التفرعات لشجرة البحث (يمكن أخذهُ بالمتوسط الحسابي لمختلف التفرعات في كل مستويات شجرة البحث ) مرفوعاً لأس العمق لشجرة البحث (عدد المستويات)
- المساحة المطلوبة من الذاكرة لخوارزمية minimax =  عدد التفرعات لشجرة البحث (يمكن أخذهُ بالمتوسط الحسابي لمختلف التفرعات في كل مستويات شجرة البحث ) مرفوعاً لأس العمق لشجرة البحث (عدد المستويات)


تعمل خوارزمية ال Alpha-Beta بنفس طريقة عمل خوارزمية minimax ولكن تحسن منه بحيث انه يتم بتر جزء من شجرة البحث وعدم البحث فيه , حيث انه لا أهمية له ولن يتم اختيارة ويمكن معرفة ذلك عن طريق متغيرين لهذه الخوارزمية 
- Alpha : وهو اقل قيمة لعقدة ال max 
- Beta : وهو اكبر قيمة لعقدة ال Min

يتم قطع شجرة البحث عند  نقاط ال Min , max اذا تحقق الشرط (Alpha>=Beta)

وتكون للمساحة المطلوبة من الذاكرة لهذه الخوارزمية حالتين وهي : 
- اسوء حالة : اذا لم يتم قطع اي جزء من الشجرة فيكون مساوي للمساحة المطلوبة من خوارزمية minimax
- افضل حالة : عندما تكون افضل حركة لكل لاعب هي الحركة الأولى فسيتم بتر مابعدها وذلك يعطي مجالا بالبحث مرتين عمقا .
 - مساحة المطلوبة من الذاكرة = عدد تفرعات شجرة البحث (يمكن اخذه بالمتوسط الحسابي لمختلف التفعات في كل مستويات شجرة البحث) مرفوعا لأس العمق لشجرة البحث مقسوماً على 2 (2/عدد المستويات )

```python

class Vertex:

    def __init__(self, name, neighbors=[], parent=None ):
        self.name = name      
        self.neighbors=neighbors.copy()        
        self.parent=parent
        
        
    def IsInNeighborsList(self, node):
        for x in self.neighbors:
            if node==x.name:
                return True
            return False     
```

```python
class Game:
    
    def __init__(self):
        self.initialize_game()

    def initialize_game(self):
        self.current_state = [[0,5,0,0,0,5],
                             [5,0,5,0,0,0],
                             [0,5,0,5,5,0],
                             [0,0,5,0,5,0],
                             [0,0,5,5,0,5],
                             [5,0,0,0,5,0]]
        
        self.playerTurn='short'
        

           
    
    vertices = {}
    vertex_indices = {} 
    start=None
    goal=None
    visited=[] 
    
    
    
    def draw_current_state(self): # [2]
        for i in range(0, 6):
            for j in range(0, 6):
                print('{}'.format(self.current_state[i][j]), end=" ")
            print()
        print()
  
   
    
  
    def add_vertex(self, vertex):
        if isinstance(vertex, Vertex) and vertex.name not in self.vertices: 
            self.vertices[vertex.name] = vertex 
            self.vertex_indices[vertex.name] = len(self.vertex_indices) 
            return True
        else:
            return False


            
    def get_key(self,val): 
        for key, value in self.vertex_indices.items(): 
            if val == value: 
                return key

    def find_neighbors(self,node_name):
        i=0
        while i<len(self.current_state):
            if self.current_state[self.vertex_indices[node_name]][i]>0:
                child_name=self.get_key(i)
                if not self.vertices[node_name].IsInNeighborsList(child_name):             
                    self.vertices[node_name].neighbors.append(self.vertices[child_name])
            i=i+1
            

    def initilzingPathCount(self,s,d): # *ملاحظة
        visited=[False]*len(current_state)
        pathCount=[0]
        CalculatePossiblePaths(s,d,visited,pathCount)
        return pathCount[0]
    
    def CalculatePossiblePaths(self,s,d,visited,pathCountNumber):# *ملاحظة
  
        visited[self.vertex_indices[s.name]]=True
    
        if(s.name==d.name):
            pathCountNumber[0] += 1
        
        else:
            i=0
            while i<len(s.neighbors):
                ineighbor=s.neighbors[i]
                if(not visited[self.vertex_indices[s.neighbors[i].name]]):
                    CalculatePossiblePaths(s.neighbors[i],d,visited,pathCountNumber)                
                i=i+1
            
        visited[self.vertex_indices[s.name]]=False

        
        
    def initilizeIstherePath(self,x,y):
        visited=[False]*len(self.current_state) 
        result=False
        return self.IstherePath(x,y,visited,result)
        
    def IstherePath(self,s,d,visited,result):  
        visited[self.vertex_indices[s.name]]=True
        
        if(s.name==d.name):
            return True
        
        else:
            for i in range(0,6):
                if not visited[self.vertex_indices[self.get_key(i)]] and self.current_state[self.vertex_indices[s.name]][i]==4:
                    nextNode= self.vertices[self.get_key(i)]
                    result= self.IstherePath(nextNode,d,visited,result)
            return result
                
                    
                
            visited[self.vertex_indices[s.name]]=False 
            
        
        
        
        
        
    def initilizeIsthereNoPath(self,x,y):
        visited=[False]*len(self.current_state)
        result= True
        return self.IsthereNoPath(x,y,visited,result)
        
     
    def IsthereNoPath(self,s,d,visited,result):
        
        visited[self.vertex_indices[s.name]]=True
        
        if(s.name==d.name):
            return False
        
        else:
            
            for i in range(0,6):
                if not visited[self.vertex_indices[self.get_key(i)]] and (self.current_state[self.vertex_indices[s.name]][i]==4 or self.current_state[self.vertex_indices[s.name]][i]==5):
                    nextNode= self.vertices[self.get_key(i)]
                    result= self.IsthereNoPath(nextNode,d,visited,result)
                    
            return result
                  
                    
                
            visited[self.vertex_indices[s.name]]=False
            
                  

    def is_end(self): #[2]
        
        if(self.initilizeIstherePath(self.start,self.goal)==True):
            return 'short'
        
        elif(self.initilizeIsthereNoPath(self.start,self.goal)==True):
            return 'cut'
        
        else:
            return None
        
        
        
        
    def is_valid(self,px,py): #[2]
        
        if px<0 or px>5 or py<0 or py>5:
            return False
        
        elif self.current_state[px][py]==0 or self.current_state[px][py]==-3 or self.current_state[px][py]==4:
            return False
           
        
        else: 
            return True
    
    
    def max_alpha_beta(self, alpha, beta): #[2]
        
        maxv = -2
        px = None
        py = None

        result = self.is_end()

        if result == 'short':
            return (-1, 0, 0)
        elif result == 'cut':
            return (1, 0, 0)
        

        for i in range(0, 6):
            for j in range(0, 6):
                if self.current_state[i][j] == 5 and self.current_state[j][i]==5:
                    self.current_state[i][j] = -3
                    self.current_state[j][i] = -3
                    (m, min_i, in_j) = self.min_alpha_beta(alpha, beta)
                    if m > maxv:
                        maxv = m
                        px = i
                        py = j
                    self.current_state[i][j] = 5
                    self.current_state[j][i] = 5

                   
                    if maxv >= beta:
                        return (maxv, px, py)

                    if maxv > alpha:
                        alpha = maxv

        return (maxv, px, py)
    
    
    def min_alpha_beta(self, alpha, beta): #[2]

        minv = 2

        qx = None
        qy = None

        result = self.is_end()

        if result == 'short':
            return (-1, 0, 0)
        elif result == 'cut':
            return (1, 0, 0)
        

        for i in range(0, 6):
            for j in range(0, 6):
                if self.current_state[i][j] == 5 and self.current_state[j][i]==5:
                    self.current_state[i][j] = 4
                    self.current_state[j][i] = 4
                    (m, max_i, max_j) = self.max_alpha_beta(alpha, beta)
                    if m < minv:
                        minv = m
                        qx = i
                        qy = j
                    self.current_state[i][j] = 5
                    self.current_state[j][i] = 5

                    if minv <= alpha:
                        return (minv, qx, qy)

                    if minv < beta:
                        beta = minv

        return (minv, qx, qy)
    
    
    
    
    def play_alpha_beta(self): #[2]
        
        while True:
            self.draw_current_state()
            self.result = self.is_end()
            
            
            if self.result != None:
                if self.result == 'short':
                    print('The winner is short!')
                elif self.result == 'cut':
                    print('The winner is cut!')


                self.initialize_game()
                return
        
            if self.playerTurn == 'short':

                while True:
                
                    (m, qx, qy) = self.min_alpha_beta(-2, 2)
                    print('Recommended move: X = {}, Y = {}'.format(qx, qy))

                    px = int(input('Insert the X coordinate: '))
                    py = int(input('Insert the Y coordinate: '))

                    qx = px
                    qy = py

                    if self.is_valid(px, py):
                        self.current_state[px][py] = 4
                        self.current_state[py][px] = 4
                        self.playerTurn = 'cut'
                        break
                    else:
                        print('The move is not valid! Try again.')
                    
                    
            else:
                (m, px, py) = self.max_alpha_beta(-2, 2)
                self.current_state[px][py] = -3
                self.current_state[py][px] = -3
                self.playerTurn = 'short'
            
            
            
            
                
          
```
